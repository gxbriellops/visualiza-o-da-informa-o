```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Credit Risk Benchmark Dataset.csv')
```

```python
df.head()
```

### Plot 1 - Gráfico de barras -  

```python
plt.figure(figsize=(10, 8))
sns.countplot(data=df, x='open_credit')
plt.title('Quantidade de linhas de crédito abertas por clientes')
plt.xlabel('Linhas de crédito abertas')
plt.ylabel('Quantidade de clientes')
plt.xticks(rotation=50)
plt.tight_layout()
plt.show()
```

### Plot 2 - Gráfico de Linhas - 

```python
plt.figure(figsize=(10, 8))
sns.lineplot(data=df, x='age', y='monthly_inc', errorbar=None)
plt.title('Evolução da renda mensal em relação a idade')
plt.xlabel('Idade')
plt.ylabel('Renda')
plt.tight_layout()
plt.show()
```

### Plot 3 - Gráfico de setores - 

```python
dependents_counts = df['dependents'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(x=dependents_counts, labels=dependents_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição de clientes por número de dependentes')
plt.axis('equal')
plt.show()
```
